```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


driver = webdriver.Chrome()
data_matrix = []
# Navigate to the desired website
for i in range (1, 5):
    url = 'https://www.sociolla.com/139-makeup?page=' + str(i)
    driver.get(url)

    # Wait until the desired elements are present
    wait = WebDriverWait(driver, 20)  # Wait for a maximum of 20 seconds
    product_divs = wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME, 'loaded-item')))

    page_data = []
    for product_div in product_divs:
        try:
            brand = product_div.find_elements(By.CLASS_NAME, 'product__brand')
            name = product_div.find_elements(By.CLASS_NAME, 'product__name')
            name = name[0].find_elements(By.TAG_NAME, 'a')
            price = product_div.find_elements(By.CLASS_NAME, 'product__price')
            review = product_div.find_elements(By.CLASS_NAME, 'product__reviews')

            page_data.append({
                    "brand": brand[0].text,
                    "price": price[0].text,
                    "name": name[0].text,
                    "review": review[0].text
            })
            
        except :
            driver.quit()

    data_matrix.extend(page_data)

driver.quit()
```

```python
import pandas as pd
df = pd.DataFrame(data_matrix)
df
```
